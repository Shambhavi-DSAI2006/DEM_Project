Data Cleaning

```python
# Create a copy

clean_df = df.copy()

# Missing values before cleaning

print("Missing values BEFORE cleaning:")
print(clean_df.isnull().sum())
```

```python
# Fill missing values using median

columns_to_fill = [
    "at_c",
    "rh_percent",
    "ws_m_s",
    "wd_deg",
    "rf_mm",
    "tot_rf_mm",
    "sr_w_mt2",
    "bp_mmhg",
    "vws_m_s"
]

for col in columns_to_fill:
    clean_df[col] = clean_df[col].fillna(
        clean_df[col].median()
    )
```

```python
print("Missing values AFTER cleaning:")
print(clean_df.isnull().sum())
```

```python
clean_df.isnull().sum()
```

Data Transformation

```python
clean_df["datetime"] = pd.to_datetime(clean_df["datetime"])

clean_df["weekday"] = clean_df["datetime"].dt.day_name()

clean_df["is_weekend"] = clean_df["weekday"].isin(
    ["Saturday", "Sunday"]
).astype(int)

clean_df[["datetime", "weekday", "is_weekend"]].head()
```

```python
clean_df["quarter"] = clean_df["datetime"].dt.quarter

clean_df[["month", "quarter"]].head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8, 9]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

clean_df["season"] = clean_df["month"].apply(get_season)

clean_df["season"].value_counts()
```

# Pollution Categorization

```python
def pollution_category(value):
    if value <= 50:
        return "Good"
    elif value <= 100:
        return "Moderate"
    elif value <= 200:
        return "Poor"
    elif value <= 300:
        return "Very Poor"
    else:
        return "Severe"

clean_df["pollution_level"] = clean_df["value"].apply(
    pollution_category
)

clean_df["pollution_level"].value_counts()
```
